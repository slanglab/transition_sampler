```python
#histogram
stem = 'train_nsamp10'
fin = 'feats/'+stem + '_pred2numexs.json'
pred2numexs = json.load(open(fin, 'r'))
print pred2numexs
```

```python
#count how many 
ex2count = Counter()
for k, v in pred2numexs.iteritems(): 
    ex2count[v] += 1
ex2count = dict(ex2count)
print ex2count
```

```python
plt.bar(ex2count.keys(), ex2count.values(), 1000)
plt.xlabel('number of examples for one predicate')
plt.ylabel('count')
plt.show()
```

```python
#stats
min_exs = min(ex2count.keys())
max_exs = max(ex2count.keys())
median_exs = np.median(np.array(ex2count.keys()))
print 'MIN num exs per one predictate={0}, count={1}'.format(min_exs, ex2count[min_exs])
print 'MAX num exs per one predictate={0}, count={1}'.format(max_exs, ex2count[max_exs])
print 'MEDIAN num exs per one predictate={0}, count={1}'.format(median_exs, ex2count[median_exs])
```
